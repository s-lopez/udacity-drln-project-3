```python
from ounoise import OUNoise
from matplotlib import pyplot as plt
import numpy as np
```

```python
n = OUNoise(1, mu=0, theta=1, sigma=.15, decay_rate=.95)
N = 1000
x = np.arange(0, N)
y = []
for i in range(N):
    new = n.sample()
    n.decay_step()
    y.append(new[0])
plt.plot(x, y)
```

```python
n = OUNoise(1, mu=0, theta=1, sigma=.15, decay_rate=.995)
N = 1000
x = np.arange(0, N)
y = []
for i in range(N):
    new = n.sample()
    n.decay_step()
    y.append(new[0])
plt.plot(x, y)
```

```python
n = OUNoise(1, mu=.0, theta=1, sigma=.15, decay_rate=.95)
episodes = 100
steps_per_episode = 100
x = np.arange(0, episodes*steps_per_episode)
y = []
for e in range(episodes):
    n.decay_step()
    for i in range(steps_per_episode):
        new = n.sample()
        y.append(new[0])
plt.plot(x, y)
```
